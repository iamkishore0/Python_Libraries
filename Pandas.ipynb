***LOC and iLOC***

```python
#LOC - accessing rows
df.loc[1]
```

```python
#Disadvantage with LOC - STOP also included
df.loc[0:5]
```

```python
#Rectify the disadvantage of LOC with iLOC
df.iloc[0:5]
```

```python
#loc with [start:start,'column name']

df.loc[0:5,'Deaths']
```

```python
#Range not only for rows and also for columns
df.loc[0:5,'Deaths':'New recovered']
```

***Handling Missing Data***

```python
#Creating a dataframe with df2
my_dict = {'First Score':[100, 90, np.nan, 95],
        'Second Score': [30, 45, 56, np.nan],
        'Third Score':[np.nan, 40, 80, 98]}

df2 = pd.DataFrame(my_dict)
df2
```

```python
#Filling the missing data with default values, we can give int,float, str and True(any)
df2.fillna(40)
```

```python
#Fill missing data with slicing
df2['Third Score'].fillna(20)
```

```python
#Drop missing data
df2.dropna()
```

***Data filtering and conditional changes***

```python
#Simple condition
df.loc[df['Active']<50]
```

```python
#Muti-conditional
df.loc[(df['Deaths']>10) & (df['Deaths']<50)]
```

```python
#Filtering strings with containing letter
df.loc[df['Country/Region'].str.contains('b')]
```

```python
#Filtering strigs with starting letters
df.loc[df['Country/Region'].str.startswith('I')]
```

```python
#Recomending safe to travel regions with multi conditions

df['safe to travel']=0
df.loc[df['New cases']<20,['safe to travel']]='yes'
df.loc[(df['New cases']>=20) & (df['New cases']<40), ['safe to travel']]='under observation'
df.loc[df['New cases']>40,['safe to travel']]='no'
```
